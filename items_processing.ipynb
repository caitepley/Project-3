```python
loc_df.drop(columns=['wdl_id','Creator','Delineator', 'Author', 'Surveyor', 'Compiler',
       'Translator', 'Contributor', 'Engraver', 'Cinematographer', 'Director',
       'Illustrator', 'Sponsor', 'Draftsman', 'Editor', 'Corrector',
       'Responsible Party', 'Calligrapher', 'Interviewee', 'Interviewer',
       'Associated Name', 'Printer', 'Lithographer', 'Donor',
       'Recording Engineer', 'Musician', 'Composer', 'Architect', 'References',
       'Attributed Name', 'Dedicatee', 'Signer', 'Copyright Holder',
       'Collector', 'Printmaker', 'Copyright claimant', 'Performer', 'Owner',
       'Commentator', 'Etcher', 'Wood-engraver', 'Scribe',
       'Author of Introduction, etc.', 'Author of Afterword, Colophon, etc.',
       'Scientific advisor', 'Speaker', 'Commentator for Written Text',
       'Illuminator', 'Librettist', 'Client', 'Transcriber', 'Collaborator',
       'Producer', 'Videographer', 'Vocalist', 'Patron', 'Funder', 'Annotator',
       'Woodcutter', 'Designer', 'Author in Quotations or Text Extracts',
       'Lyricist', 'Type Designer', 'Animator', 'Writer of added text',
       'Project Director', 'Dubious Author', 'Censor', 'Engineer', 'Lead',
       'Publishing Director', 'Singer', 'Binder', 'Originator', 'Typographer',
       'Book producer', 'Narrator', 'Colorist', 'Arranger', 'Recipient',
       'Correspondent','Creator','Language', 'Artist',
       'Cartographer', 'Note','Photographer','Place','Subject Date','Additional Subjects','description','Physical Description'], inplace=True)
loc_df.head()
```

```python
loc_df.head()
```

```python
loc_df['Dewey Decimal Code'] = loc_df['Dewey Decimal Code'].replace('\|', '.', regex=True)
loc_df.head()
```

```python
item_type_df = pd.read_csv('Resources/item_codes_new.csv', encoding='utf-8')
item_type_df
```

```python
items_df = loc_df.join(item_type_df,how='outer')
items_df.drop(columns=['Type of Item','Item'],inplace=True)
items_df.head()
```

```python
col_type_df = pd.read_csv('Resources/col_mapping_new.csv', encoding='utf-8')
col_type_df.head()
```

```python
items_df = items_df.merge(col_type_df,how='outer')
items_df.drop(columns='collection',inplace=True)
items_df.head()
```

```python
ins_type_df = pd.read_csv('Resources/ins_mapping_new.csv', encoding='utf-8')
ins_type_df.head()
```

```python
items_df = items_df.merge(ins_type_df,how='outer')
items_df.drop(columns='Institution',inplace=True)
items_df.head()
```

```python
pub_type_df = pd.read_csv('Resources/pub_mapping_new.csv', encoding='utf-8')
pub_type_df.head()
```

```python
items_df = items_df.merge(pub_type_df,how='outer')
items_df.drop(columns='Publisher',inplace=True)
items_df.head()
```

```python
items_df = items_df.loc[:,['title','col', 'wdl_url','ins_key','pub_key','Dewey Decimal Code','item_code','Date Created']]
```

```python
items_df.head()
```

```python
items_df['item_id'] = items_df.index
items_df.head()
```

```python
# Export the mapping DataFrame to a CSV file
items_df.to_csv('Resources/items.csv', index=False)
```
