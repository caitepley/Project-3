```python
#Isolate the Type of Item for the Item_Type Table


item_type = loc2_df['Type of Item'].unique()
print(item_type)
```

```python
# Define a dictionary to map item types to codes
type_code_mapping = {
    'Prints, Photographs': 'prph',
    'Maps': 'mp',
    'Books': 'bk',
    'Manuscripts': 'ms',
    'Motion Pictures': 'mp',
    'Sound Recordings': 'sr',
    'Journals': 'jr',
    'Newspapers': 'np'
}

# Add a new column 'item_type' based on the mapping
loc2_df['item_type'] = loc2_df['Type of Item'].map(type_code_mapping)

# Print the updated DataFrame with the Item Code. Later remove the Type of Item column - verifying that item code was mapped
loc2_df
```

Creating the Item Code csv file for importing into Database

```python
# Create a DataFrame with unique item codes and their corresponding item types
unique_item_code_df = pd.DataFrame(list(type_code_mapping.items()), columns=['Type of Item', 'Item Code'])

# Export to CSV file
unique_item_code_df.to_csv('unique_item_codes.csv', index=False)

```

Creating an isolated dataframe for the collections table mapping codes

```python
# Step 1: Isolate the column and handle NaN values
isolated_column = loc2_df['collection'].fillna('Non-Collection')

# Step 2: Extract unique values
unique_values = isolated_column.unique()

# Step 3: Create a DataFrame for the mapping with a prefixed col_ key
prefix = 'col'
col_mapping_df = pd.DataFrame({'collection': unique_values, 'col': [f'{prefix}_{i}' for i in range(1, len(unique_values) + 1)]})

# Step 4: Map the primary key to the main dataframe
loc2_df = loc2_df.merge(col_mapping_df, on='collection', how='left')

# Display the result
loc2_df
```

```python
# Export the mapping DataFrame to a CSV file
col_mapping_df.to_csv('col_mapping_file.csv', index=False)
```

Creation of Institution dataframe, mapping code and exporting as csv file

```python
# Step 1: Isolate the column and handle NaN values
instit_column = loc2_df['Institution'].fillna('unknown')

# Step 2: Extract unique values
instit_values = instit_column.unique()

# Step 3: Create a DataFrame for the mapping with a prefixed primary key
prefix = 'ins'
ins_mapping_df = pd.DataFrame({'Institution': instit_values, 'ins_key': [f'{prefix}_{i}' for i in range(1, len(instit_values) + 1)]})

#lang_df_col5['item_id'] = lang_df_col5.index

# Step 4: Map the primary key to the main dataframe
loc2_df = loc2_df.merge(ins_mapping_df, on='Institution', how='left')

# Display the result
print(loc2_df)
```

```python
#Exporting the institution mapping file

# Export the mapping DataFrame to a CSV file
ins_mapping_df.to_csv('ins_mapping_file.csv', index=False)
```

Creation of Publisher dataframe, mapping code and exporting as csv file

```python
# Step 1: Isolate the column and handle NaN values
publish_column = loc2_df['Publisher'].fillna('unknown')

# Step 2: Extract unique values
publish_values = publish_column.unique()

# Step 3: Create a DataFrame for the mapping with a prefixed primary key
prefix = 'pub'
pub_mapping_df = pd.DataFrame({'Publisher': publish_values, 'pub_key': [f'{prefix}_{i}' for i in range(1, len(publish_values) + 1)]})


# Step 4: Map the primary key to the main dataframe
loc2_df = loc2_df.merge(pub_mapping_df, on='Publisher', how='left')

```

```python
# Export the mapping DataFrame to a CSV file
pub_mapping_df.to_csv('pub_mapping_file.csv', index=False)
```
