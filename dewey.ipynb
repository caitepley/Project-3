```python
import pandas as pd
import dask.dataframe as dd
from pathlib import Path
from sqlalchemy import create_engine, text
```

```python
#Reading in the source file for the project
file_path = 'Resources/wdl_data_en_reduced.csv'
# Specifying encoding as'ISO-8859-1' to read file
loc_df = pd.read_csv(file_path, encoding='utf-8')

#Prints the Dask DataFrame structure in which it prints the column headings and datatypes 
loc_df['Place'].unique()
```

```python
# all of the places
dewey_df = pd.DataFrame(loc_df['Dewey Decimal Code'])
dewey_df
```

```python
dewey_df = dewey_df.dropna()
dewey_df
```

```python
dewey_df = dewey_df.drop_duplicates()
dewey_df
```

```python
dewey_df = dewey_df.replace('\|', '.', regex=True)
dewey_df
```

```python
# set 'genre' column to string value depending on what the dewey decimal code is

# 000 Computer science, information & general works
# 100 Philosophy & psychology
# 200 Religion
# 300 Social sciences
# 400 Language
# 500 Science
# 600 Technology
# 700 Arts & recreation
# 800 Literature
# 900 History & geography
```

```python
dewey_df['genre'] = ''
dewey_df.head()
```

```python
dewey_df['Dewey Decimal Code'].size
```

```python
for i in range(dewey_df['Dewey Decimal Code'].size):
    if(dewey_df.iloc[i,0] < '100'):
        dewey_df.iloc[i,1] = 'Computer science, information & general work'
    if(dewey_df.iloc[i,0] < '200'):
        dewey_df.iloc[i,1] = 'Philosophy & psychology'
    if(dewey_df.iloc[i,0] < '300'):
        dewey_df.iloc[i,1] = 'Religion'
    if(dewey_df.iloc[i,0] < '400'):
        dewey_df.iloc[i,1] = 'Social sciences'
    if(dewey_df.iloc[i,0] < '500'):
        dewey_df.iloc[i,1] = 'Language'
    if(dewey_df.iloc[i,0] < '600'):
        dewey_df.iloc[i,1] = 'Science'
    if(dewey_df.iloc[i,0] < '700'):
        dewey_df.iloc[i,1] = 'Technology'
    if(dewey_df.iloc[i,0] < '800'):
        dewey_df.iloc[i,1] = 'Arts & recreation'
    if(dewey_df.iloc[i,0] < '900'):
        dewey_df.iloc[i,1] = 'Literature'
    else:
        dewey_df.iloc[i,1] = 'History & geography'
    
    
```

```python
# Export final language dataframe to CSV file
dewey_df.to_csv('Resources/Dewey.csv', index=False)
```
